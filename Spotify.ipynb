<h1>Top Tracks Analysis</h1>
<p> By analyzing the popularity scores of tracks and albums, music labels and artists can identify their most successful and popular releases. This insight can help them focus on creating similar content or promoting those tracks that have resonated well with the audience.
<p>Popularity analysis can reveal trends in music consumption, helping businesses understand what types of music are currently in demand and gaining traction among listeners. This information can be crucial for music production companies to align their offerings with current trends.


```python
# Top Tracks based on Popularity
top_tracks = df.nlargest(10, 'Popularity')[['Track Name', 'Artist Name(s)', 'Popularity']]
top_tracks
```

<h1> Genre analysis</h1>

```python
# Count the occurrences of each genre in the 'Artist Genres' column
genre_counts = df['Artist Genres'].str.split(', ').explode().value_counts()
genre_counts
```

```python
top_genres = genre_counts.head(10)
top_genres
```

```python
top_genres = genre_counts.head(10)
# Plot the top genres using a bar chart with Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x=top_genres.index, y=top_genres.values, palette='Blues')
plt.title('Top 10 Genres in the Spotify Dataset')
plt.xlabel('Genre')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

<p>It is observed that POP music is most popular, also other kinds of pop music is also famous across the world

<h1>Track Duration Analysis</h1>

```python
# Descriptive statistics of track duration
track_duration_stats = df['Track Duration (ms)'].describe()

# Histogram of track duration
plt.figure(figsize=(10, 6))
plt.hist(df['Track Duration (ms)'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Track Duration')
plt.xlabel('Track Duration (ms)')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()
```

<h1>Popularity vs. Audio Features </h1>

```python
# Select the columns for analysis
selected_columns = ['Danceability', 'Energy', 'Acousticness', 'Valence', 'Popularity']

# Create a correlation matrix
correlation_matrix = df[selected_columns].corr()

# Plot the correlation matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0, fmt='.2f')
plt.title('Correlation Matrix: Popularity vs. Audio Features')
plt.show()
```

<p> The heatmap shows the correlation coefficients between popularity and audio features. Positive values indicate a positive correlation (increasing popularity with increasing feature value), while negative values indicate a negative correlation (decreasing popularity with increasing feature value).

<h1>Analysis of Popular Artists and Their Genres.

```python
# Group by 'Artist Name(s)' and calculate the average popularity for each artist
artist_popularity = df.groupby('Artist Name(s)')['Popularity'].mean().reset_index()

# Select the top 10 popular artists
top_10_artists = artist_popularity.nlargest(10, 'Popularity')

# Merge the top 10 popular artists with the original DataFrame to get their genres
top_10_artists_genres = df[df['Artist Name(s)'].isin(top_10_artists['Artist Name(s)'])]

# Split the multiple genres in 'Artist Genres' column and explode them into separate rows
top_10_artists_genres['Artist Genres'] = top_10_artists_genres['Artist Genres'].str.split(', ')
top_10_artists_genres = top_10_artists_genres.explode('Artist Genres')

# Count the occurrences of each genre for the top 10 popular artists
genre_counts = top_10_artists_genres['Artist Genres'].value_counts().reset_index()
genre_counts.columns = ['Genre', 'Count']

# Plot the popular genres for the top 10 artists using a bar plot
plt.figure(figsize=(10, 6))
sns.barplot(x='Count', y='Genre', data=genre_counts, palette='muted')
plt.title('Popular Genres for Top 10 Popular Artists')
plt.xlabel('Number of Tracks')
plt.ylabel('Genre')
plt.tight_layout()
plt.show()


```

<h1>Track Popularity by Explicitness</h1>

```python
# Box plot of Track Popularity by Explicitness
plt.figure(figsize=(8, 6))
sns.boxplot(x='Explicit', y='Popularity', data=df, palette='Set2')
plt.title('Track Popularity by Explicitness')
plt.xlabel('Explicit')
plt.ylabel('Popularity')
plt.xticks(ticks=[0, 1], labels=['Not Explicit', 'Explicit'])
plt.tight_layout()
plt.show()
```

<h1>Popularity Distribution by Album Artist.

```python

# Group by 'Album Artist Name(s)' and calculate the average popularity for each album artist
album_artist_popularity = df.groupby('Album Artist Name(s)')['Popularity'].mean().reset_index()

# Select the top 10 album artists with the highest average popularity
top_10_album_artists = album_artist_popularity.nlargest(10, 'Popularity')

# Merge the top 10 album artists with the original DataFrame to get their tracks
top_10_album_artists_tracks = df[df['Album Artist Name(s)'].isin(top_10_album_artists['Album Artist Name(s)'])]

# Box plot of Track Popularity by Top 10 Album Artists
plt.figure(figsize=(12, 6))
sns.boxplot(x='Album Artist Name(s)', y='Popularity', data=top_10_album_artists_tracks, palette='Set3')
plt.title('Popularity Distribution by Top 10 Album Artists')
plt.xlabel('Album Artist Name')
plt.ylabel('Popularity')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

<p>It is obvserved that MORGAN WALLEN has most popular albums available
